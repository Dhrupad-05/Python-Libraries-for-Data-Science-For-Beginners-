# **Styling**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
tip= sns.load_dataset("tips")
tip
````

````python
#setstyle allows to change the background of the graph
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=tip)
plt.show()
````

````python
#despine revomes the outer boundary lines
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=tip)
sns.despine()
plt.show()
````

````python
#scaling
sns.set_style("whitegrid")
plt.figure(figsize=(12,5))
sns.barplot(x="day",y="total_bill",data=tip)
sns.despine()
plt.show()
````

````python
#context changes
sns.set_style("whitegrid")
sns.set_context("poster",font_scale=1)
sns.barplot(x="day",y="total_bill",data=tip)

plt.show()
````

````python
#context changes
sns.set_style("whitegrid")
sns.set_context("paper",font_scale=1)
sns.barplot(x="day",y="total_bill",data=tip,palette="cool")

plt.show()
````

# **FacetGrid**

````python
fg=sns.FacetGrid(tip)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
````

````python
fg=sns.FacetGrid(tip,col="time",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
````

````python
fg=sns.FacetGrid(tip,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
````

````python
fg=sns.FacetGrid(tip,col="time",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
````

Color Changes

````python
#using edgecolor
fg=sns.FacetGrid(tip,col="time",hue="day")
fg.map(plt.bar,"total_bill","tip",edgecolor="g").add_legend()
plt.show()
````

````python
#using palette
fg=sns.FacetGrid(tip,col="time",hue="day",palette="cool")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
````

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data={
    'customer':['riya','aman','faizan','neha','imran', 'sneha'],
    'age': [20,30,40,22,38,25],
    'spending': [100,200,300,110,290,130]
    }
df=pd.DataFrame(data)
x=df[['age','spending']]
model= KMeans(n_clusters=2,random_state=42,n_init=10) #n_init runs the same thing (here 10 times) to avoid bad grouping
df['Group']=model.fit_predict(x)
print(df)
plt.figure(figsize=(6,5))
for group in df['Group'].unique():
    group_data=df[df['Group']==group]
    plt.scatter(group_data['age'],group_data['spending'],label=f'Group{group}')

plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('CustomerSegments(K-Means)')
plt.legend()
plt.show()
```

```python
#pca help to remove less important data
#without it training is slower
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


data={
    'age':[25,30,35,40,45,50],
    'income':[50000,60000,70000,80000,90000,100000],
    'spending':[70,60,50,40,30,20],
    'savings':[100,5000,8000,10000,15000,20000]
}

df=pd.DataFrame(data)

scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)

pca= PCA(n_components=2)
pca_result=pca.fit_transform(df)

pca_df=pd.DataFrame(pca_result,columns=['PC1','PC2'])
print(pca_df)

explained_variance=pca.explained_variance_ratio_
print("Variance Captured by each PCA Component")
print(np.round(explained_variance*100,2))

#scatterplot
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'],pca_df['PC2'],color='black',s=80)
plt.xlabel('PCA1 Main Pattern')
plt.ylabel('PCA2 Main Pattern')
plt.title('PCA Projections(2D View)')
plt.show()
```

```python
#Step->3
#Feature Scaling
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Re-read the CSV and apply label encoding to ensure df is in the correct state
# This step is added to prevent issues from previous in-place modifications of df
df = pd.read_csv("student_management_dataset.csv")
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Ethnicity'] = le.fit_transform(df['Ethnicity'])
df['Socioeconomic_Status'] = le.fit_transform(df['Socioeconomic_Status'])
df['Parental_Education_Level'] = le.fit_transform(df['Parental_Education_Level'])
df['Past_Academic_Performance'] = le.fit_transform(df['Past_Academic_Performance'])

# Define features for the model training and prediction, matching the user input prompts.
# The previous cell (3j5YRL9cg2sK) handles label encoding, so `df` here is already processed for categorical data.
model_features_for_training_and_prediction = ['GPA','Attendance_Rate','Number_of_Late_Submissions','Class_Participation_Score',
'Online_Learning_Hours','Library_Usage_Hours']

# Extract these features from the raw DataFrame for scaling specific to the model
x_raw = df[model_features_for_training_and_prediction]
# Convert 'Current_Semester_Performance' into a binary target variable (e.g., 0 for Fail, 1 for Pass)
# Assuming a passing score threshold of 70 for demonstration purposes.
y = (df['Current_Semester_Performance'] >= 70).astype(int) # Target variable converted to binary classes

# Create a dedicated scaler for the model's input features
scaler_model = StandardScaler()
x = scaler_model.fit_transform(x_raw) # Scale x_raw to get x

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

print("Classification Report")
print(classification_report(y_test,y_pred))

print("Confusion Matrix")
# Corrected typo in variable name to avoid conflict with function name and ensure consistency
conf_matrix = confusion_matrix(y_test,y_pred)
print(conf_matrix)

plt.figure(figsize=(6,4))
# Used the corrected variable name for the heatmap
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues',
            xticklabels=[ "Fail","Pass"],yticklabels=["Fail","Pass"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout() # Added missing parentheses
plt.show()

print("Predict your Results")
try:
  GPA=float(input("Enter GPA:"))
  Attendance_Rate=float(input("Enter Attendance Rate:")) # Corrected typo
  Number_of_Late_Submissions=float(input("Enter Late Submissions:"))
  Class_Participation_Score=float(input("Enter Class Participation Score:"))
  Online_Learning_Hours=float(input("Enter Online Learning Hours:"))
  Library_Usage_Hours=float(input("Enter Library Usage Hours:"))

  user_input_df=pd.DataFrame({
      'GPA':[GPA],
      'Attendance_Rate':[Attendance_Rate], # Corrected typo
      'Number_of_Late_Submissions':[Number_of_Late_Submissions],
      'Class_Participation_Score':[Class_Participation_Score],
      'Online_Learning_Hours':[Online_Learning_Hours],
      'Library_Usage_Hours':[Library_Usage_Hours]
  })

  # Transform user input using the scaler fitted on the model's input features
  user_input_scaled=scaler_model.transform(user_input_df)
  prediction=model.predict(user_input_scaled)[0]

  Final_Results= "Pass" if prediction==1 else "Fail" # Corrected indentation
  print(f"Predicted Semester Performance: {Final_Results}") # Corrected indentation
except Exception as e:
  print(f"Error: {e}")
```

# **PairPlot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tip=sns.load_dataset("tips")
tip.head(75)
```

```python
sns.pairplot(tip)
plt.show()
```

```python
sns.pairplot(tip,hue="sex")
plt.show()
```

```python
sns.pairplot(tip,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(tip,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="flare")
plt.show()
```

```python
sns.pairplot(tip,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],kind="reg")
plt.show()
```

```python
#kind types->scatter,kde,hist,reg
sns.pairplot(tip,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],kind="kde")
plt.show()
```

```python
sns.pairplot(tip,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```

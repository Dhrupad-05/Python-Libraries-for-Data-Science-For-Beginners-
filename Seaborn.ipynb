# **HeatMap**

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
var=np.linspace(1,10,20).reshape(4,5)
var
````

````python
sns.heatmap(var)
plt.show()
````

````python
data=sns.load_dataset("anagrams")
x=data.drop(columns=["attnr"],axis=1)
x
sns.heatmap(x)
````

````python
sns.heatmap(x,vmin=0,vmax=15)
````

````python
sns.heatmap(x,vmin=0,vmax=15,cmap="copper")
````

````python
#annot Parameter is used to show the numbers on the list/block in the heatmap
sns.heatmap(x,vmin=0,vmax=15,cmap="copper",annot=True)
plt.show()
````

````python
#here we will be renaming the blocks on the heatmap of a random array
var1=np.linspace(1,10,10).reshape(2,5)
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True)
````

````python
#here we will be renaming the blocks on the heatmap of a random array
#here fmt is format which is used to give what types of data is given to name the blocks of the heat map
ar=np.array([["a1","a2","a3","a4","a5"],
             ["b1","b2","b3","b4","b5"]])
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=ar,fmt="s")
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y)
plt.show()
#here we are changing the color of the txt on the heat map blocks
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=10)
plt.show()
#here we are changing the linewidth
````

````python

var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=5,linecolor="r")
plt.show()
#here we are changing the color of linewidth
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=5,linecolor="r",cbar=False)
plt.show()
#here we are removing the color bar
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=5,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
plt.show()
#here we are removing the labels
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
w=sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=5,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
w.set(xlabel="Dhrupad",ylabel="Paitandy")
plt.show()
#here we are giving labels
````

````python
var1=np.linspace(1,10,10).reshape(2,5)
y={"fontsize":20,"color":"r"}
w=sns.heatmap(var1,vmin=0,vmax=10,cmap='copper',annot=True,annot_kws=y,linewidths=5,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
w.set(xlabel="Dhrupad",ylabel="Paitandy")
sns.set(font_scale=9)
plt.show()
#here we are changing font size
````

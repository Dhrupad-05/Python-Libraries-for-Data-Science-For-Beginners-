# **Handling Missing Values**

```python
arr=np.array([1,2,np.nan,4,np.nan,6])
cleaned_arr=np.nan_to_num(arr)
cleaned_arr
```

```python
arr=np.array([1,2,np.nan,4,np.nan,6])
cleaned_arr=np.nan_to_num(arr,nan=100)#putting in 100 in place of nan
cleaned_arr
```

```python
#check wheter there is infinite value
arr=np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
```

```python
arr=np.array([1,2,np.inf,4,-np.inf,6])
cleaned_arr=np.nan_to_num(arr, posinf=1000, neginf=-1000)#passing 1000 inplace of infinite values
cleaned_arr
```

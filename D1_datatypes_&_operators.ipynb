### Key Takeaways

* **Escape Sequences**: Special characters used inside strings to control formatting.
    * `\n`: Creates a **new line** in the output.
    * `\t`: Inserts a **tab** (typically equivalent to 4 spaces).
    * They must be used inside quotation marks (e.g., `"Hello\nWorld"`).

* **Finding a Minimum Value Logically**: A clever method to find the minimum number without direct comparisons in the final step.
    * First, determine which variable is the minimum using boolean flags (e.g., `min1 = (A < B) and (A < C)`). This will result in one flag being `True` and the others `False`.
    * In Python, `True` evaluates to `1` and `False` evaluates to `0`.
    * Multiply each boolean flag by its corresponding variable: `min1 * A`, `min2 * B`, etc.
    * The term with the `True` flag will become `1 * variable` (the variable itself), while all others become `0 * variable` (zero).
    * Add the results together. Since only one term is non-zero, the sum will be the minimum value. Example: `(min1 * A) + (min2 * B) + ...`.

* **Logical `not` Operator**: Inverts a boolean value.
    * `not True` results in `False`.
    * `not False` results in `True`.
    * When applied to numbers, any non-zero value is treated as `True`, and `0` is treated as `False`. So, `not 10` is `False`, and `not 0` is `True`.

* **Number Systems**: Different ways to represent numbers. The **base** determines the number of unique digits used.
    * **Decimal (Base-10)**: Uses digits 0-9. This is the standard system for humans.
    * **Binary (Base-2)**: Uses digits 0 and 1. This is the fundamental language of computers.
    * **Octal (Base-8)**: Uses digits 0-7.
    * **Hexadecimal (Base-16)**: Uses digits 0-9 and letters A-F.
    * **Combinations**: The number of possible combinations for `n` variables (bits) in a system with base `b` is **bⁿ**. For example, 3 binary bits give 2³ = 8 combinations.

* **Bitwise Operations**: Operations that work directly on the binary digits (bits) of numbers.
    * In Python, you must prefix binary numbers with `0b` (e.g., `0b1010`).
    * The `bin()` function can be used to convert a decimal number to its binary string representation.
    * Operators include **AND (`&`)**, **OR (`|`)**, **NOT (`~`)**, **Left Shift (`<<`)**, and **Right Shift (`>>`)**.
    * **Right Shift (`>>`)**: Moves bits to the right, effectively dividing the number by 2 for each shift position.
    * **Left Shift (`<<`)**: Moves bits to the left, effectively multiplying the number by 2 for each shift position.

The session begins by revisiting a previous problem involving **bitwise operations**, explaining why a negative number input changes the output. It then introduces various number systems and the concepts of **Least Significant Bit (LSB)** and **Most Significant Bit (MSB)**, clarifying that the MSB acts as a sign bit (1 for negative, 0 for positive) in a 32-bit system. ### Key Takeaways

* **Sign Bit (MSB)**: In a 32-bit system, the leftmost bit (Most Significant Bit) represents the sign of a number: `1` for **negative** and `0` for **positive**. This is crucial for understanding how computers handle arithmetic with negative numbers.
* **Bitwise Right Shift (`>>`)**: Shifting a 32-bit number to the right 31 times effectively moves the MSB (sign bit) to the LSB's position. This is the key to solving the initial problem.
* **Derived Logical Operations**:
    * **XOR (`^`)**: Gives `1` (True) if the inputs are **different** and `0` (False) if they are the **same**. It's useful for tasks like toggling states (e.g., a two-way switch).
    * **NOR**: An `OR` operation followed by a `NOT`.
    * **NAND**: An `AND` operation followed by a `NOT`.
* **Control Structures (`if-elif-else`)**:
    * These statements allow the program to execute different blocks of code based on whether a condition is true or false.
    * The syntax requires a **colon (`:`)** at the end of the condition and proper **indentation** for the code block.
    * Conditions are checked sequentially. Once a condition is met, its block is executed, and the program skips the rest of the `elif`/`else` blocks.
* **User Input**:
    * The **`input()`** function is used to get input from the user, which is always returned as a **string**.
    * The **`eval()`** function is used to evaluate the string input and convert it into a numeric type (integer or float) automatically. This is more flexible than using `int()` or `float()` when the user input type is unknown.
* **Loops (`while` and `for`)**:
    * **`while` loop**: Repeats a block of code as long as a specified condition is `True`. You must manually manage the loop variable (e.g., incrementing it) to avoid an infinite loop.
    * **`for` loop**: Repeats a block of code for each item in a sequence (like a range of numbers). It handles the incrementing automatically.
    * The **`range(n)`** function generates a sequence of numbers from `0` up to (but not including) `n`.

* **Ternary Operator**: A concise way to write a simple `if-else` statement in a single line. It's great for making code shorter and more readable for simple assignments.

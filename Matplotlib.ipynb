````python
#creating 2 or more plots in a same window
#plt.subplot(n rows, m columns, index)
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1)#1row ,2column,1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)# 1row, 2colimn, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt


#fig, ax=plt.subplot(nrows,mcols,figsize=(width,height))
fig, ax=plt.subplots(1,2,figsize=(10,5))


x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y,color='red')
ax[0].set_title('Line Chart')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line & Bar Charts')

plt.tight_layout()
plt.show()
````

````python
movie_df = netflix[netflix['type'] == 'Movie'].copy()

# Keep only rows that contain 'min' in the duration to avoid errors during conversion
movie_df = movie_df[movie_df['duration'].str.contains('min', na=False)]

# Safely extract integer duration, specifying regex=False for literal replacement
movie_df['duration_int'] = movie_df['duration'].str.replace(' min', '', regex=False).astype(int)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration (in Mins)')
plt.ylabel('No. of Movies')
plt.tight_layout()
plt.show()
````

na=False
Does two things:

str.contains('min')

Checks each row in the "duration" column.

Returns True if the string contains "min", otherwise False.

Example:

"90 min" ‚Üí True

"2 Seasons" ‚Üí False

na=False

Deals with missing values (NaN) in the column.

By default, if a row has NaN (missing data), the result of .str.contains() would also be NaN (not True or False).

But na=False forces those NaN rows to be treated as False.

So effectively:

"90 min" ‚Üí True ‚úÖ

"2 Seasons" ‚Üí False ‚ùå

NaN (missing) ‚Üí False ‚ùå (instead of error/NaN)

Without na=False

If there are NaNs in the duration column, .str.contains() would return NaN for those rows.
When you try to filter the DataFrame using that, it might throw an error or drop them unpredictably.

**Regex** stands for **Regular Expression**.
It‚Äôs a powerful tool used for **pattern matching and text manipulation**.

Think of it like a search language for strings ‚Äî instead of just looking for exact words, regex lets you describe *patterns* of text.

---

### Simple Example (without regex)

```python
text = "90 min"
text.replace(" min", "")   # removes the exact substring " min"
```

This works only if the text is **exactly `" min"`** at the end.

---

### Example (with regex)

```python
import re
re.sub(r"\D", "", "90 min")   # -> "90"
```

Here `\D` means "any non-digit character".
So it removes letters and spaces, leaving only `"90"`.

---

### Why it matters in your Netflix dataset:

In your code:

```python
movie_df['duration'].str.replace(' min','')
```

By default, **pandas treats the first argument as regex**.

* If you pass `' min'`, pandas thinks it‚Äôs a regex pattern, not just plain text.
* In newer pandas, you can tell it **not** to use regex:

  ```python
  .str.replace(' min', '', regex=False)
  ```

---

### Quick Regex Cheatsheet:

* `\d` ‚Üí digit (0‚Äì9)
* `\D` ‚Üí non-digit
* `.` ‚Üí any character
* `^` ‚Üí start of string
* `$` ‚Üí end of string
* `+` ‚Üí one or more
* `*` ‚Üí zero or more

Example:

* Pattern: `r"^\d+ min$"`

  * `^\d+` ‚Üí starts with digits
  * ` min` ‚Üí followed by " min"
  * `$` ‚Üí ends there
  * Matches: `"90 min"`, `"120 min"`

---

üëâ So, **regex is like advanced search for strings**, which is why pandas asks whether your `.replace()` is regex or plain text.

Would you like me to show you how to use regex directly to extract **only the numbers (durations)** from the Netflix column, instead of using `.replace()`?










````python
movie_df = netflix[netflix['type'] == 'Movie'].copy()

# Keep only rows that contain 'min' in the duration
movie_df = movie_df[movie_df['duration'].str.contains('min', na=False)]

# Safely extract integer duration
movie_df['duration_int'] = movie_df['duration'].str.replace(' min', '', regex=False).astype(int)

# Plot
plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration (in mins)')
plt.ylabel('No. of Movies')
plt.tight_layout()
plt.show()


````

## üß© What Is a Factor Plot?

A **factor plot** is a categorical plot that shows the relationship between two variables, with optional grouping by categories like hue or column facets. By default, it produces a **point plot**, but can generate other plot types using the `kind` argument. ([TutorialsPoint][1])

Example usage:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.factorplot(x="day", y="total_bill", data=tips, kind="bar")
plt.show()
```

# **FactorPlot**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
````

````python
df= sns.load_dataset("tips").head(20)
df
````

````python
X=df[["total_bill","tip","smoker","day","time","size"]]
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex")
plt.show()
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex",kind="box")
plt.show()
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex",kind="violin")
plt.show()
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex",orient="h")
plt.show()
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex",palette="viridis")
plt.show()
````

````python
sns.catplot(x="size",y="tip",data=df,hue="sex",palette="viridis",height=10)
plt.show()
````

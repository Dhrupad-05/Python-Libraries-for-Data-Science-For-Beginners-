````python
import pandas as pd
import numpy as np


#here https://www.football-data.co.uk/mmz4281/2526/ is the same for all the links so it's called the "Root'
root= "https://www.football-data.co.uk/mmz4281/"

leagues= ['E1', 'E2', 'E3', 'EC']
frames= []    #empty list to store the data

#looping through the leagues, read multiple csv & append it into a list
for league in leagues:
  for season in range(15,21):
    df=pd.read_csv( root + '2526'+ '/'+ league + '.csv', encoding ='unicode_escape')
    df.insert(1,'season', season)
    frames.append(df)
````

**Organise ALL DATA in a Dictionary**

````python
dict_countries=  {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1',
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
````

````python
#Certainly! This code snippet is designed to fetch historical football data from a website, organize it, and store it in a dictionary of pandas DataFrames. Here's a step-by-step breakdown:
dict_historical_data = {}: Initializes an empty dictionary called dict_historical_data. This dictionary will eventually store the concatenated DataFrames for each league, with the league names as keys.
leagues = ['E1', 'E2', 'E3', 'EC']: Defines a list named leagues containing abbreviated codes for different football leagues. Note: While this list is defined, it's not actually used in the subsequent loops. The code iterates through the keys of dict_countries instead.
#empty list to store the data: This is a comment indicating the purpose of the next line.
#looping through the leagues, read multiple csv & append it into a list: This comment describes the main process the following nested loops will perform.
for league in dict_countries:: This is the outer loop. It iterates through the keys of the dict_countries dictionary. In each iteration, the variable league will take on the value of a league name (e.g., 'Spanish La Liga', 'German Bundesliga', etc.).
frames = []: Inside the outer loop, this line initializes an empty list called frames for each league. This list will temporarily store the DataFrames for each season within the current league.
for season in range(15, 21):: This is the inner loop. It iterates through a range of numbers from 15 up to (but not including) 21. These numbers are used to represent the seasons (e.g., 15 corresponds to the 2015-2016 season, 16 to 2016-2017, and so on, up to 20 for 2020-2021).
df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv'): This is the core line for data fetching.
root: This variable (presumably defined earlier in your notebook as '[redacted link]) is the base URL for the data.
str(season) + str(season + 1): This creates the season part of the URL (e.g., '1516', '1617', etc.).
'/': Adds a forward slash to the URL.
dict_countries[league]: This accesses the value associated with the current league key from the dict_countries dictionary (e.g., 'SP1', 'D1', etc.). This provides the abbreviated league code needed for the URL.
'.csv': Appends the file extension to complete the URL.
pd.read_csv(...): Reads the CSV file from the constructed URL into a pandas DataFrame and assigns it to the variable df.
df.insert(1, 'season', season): This line inserts a new column named 'season' into the DataFrame df at position 1 (the second column). The values in this new column are the current season number from the inner loop. This adds a column to identify the season for each row of data.
frames.append(df): Appends the DataFrame df (with the added 'season' column) to the frames list for the current league.
df_concat = pd.concat(frames): After the inner loop finishes (meaning all seasons for a specific league have been processed), this line concatenates all the DataFrames stored in the frames list into a single DataFrame called df_concat.
dict_historical_data[league] = df_concat: This line adds an entry to the dict_historical_data dictionary. The key is the current league name (e.g., 'Spanish La Liga'), and the value is the concatenated DataFrame df_concat containing data for all seasons of that league.
````

````python
dict_historical_data={}
leagues= ['E1', 'E2', 'E3', 'EC']
  #empty list to store the data

#looping through the leagues, read multiple csv & append it into a list
for league in dict_countries:
  frames= []
  for season in range(15,21):
    df=pd.read_csv( root + str(season)+str(season+1) + '/' + dict_countries[league] + '.csv')
    df.insert(1,'season', season)
    frames.append(df)

  #conatenate
  df_concat=pd.concat(frames)
  dict_historical_data[league]= df_concat
````

# **Box Plot**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip=sns.load_dataset("tips")
tip
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="sex",palette='dark:b',order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
#show mean shows the mean as the single line
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="sex",palette="dark:b",showmeans=True)
plt.show()
```

```python

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="sex",palette='dark:b',showmeans=True,

            meanprops={"marker":"+","markeredgecolor":"w"})
plt.show()
```

```python

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=tip,hue="sex",color="b",showmeans=True,linewidth=3,palette="plasma")
plt.show()
```

```python
#to show the graph horizontally we need to numeric values in the x bar
#as we don't have that we will change the x axis to y and the y to x
#show mean shows the mean as the single line
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=tip,hue="sex",color="b",showmeans=True,linewidth=3,palette="plasma")
plt.show()
```

```python
#show mean shows the mean as the single line
sns.set(style="whitegrid")
sns.boxplot(data=tip,showmeans=True,linewidth=3,palette="plasma")
plt.show()
```
